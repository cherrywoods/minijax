# Derivatives
Compute first and second derivatives using `minijax.grad`.

```python
import matplotlib.pyplot as plt
import numpy as np

from minijax.core import relu, reshape
from minijax.eval import Array
from minijax.grad import grad
from minijax.nn import init_mlp
from minijax.vmap import vmap
```

## ReLU

```python
drelu = grad(relu)
ddrelu = grad(drelu)


x = Array(np.linspace(-5, 5, 1000))
plt.plot(x.array, relu(x).array, label="ReLU(x)")
plt.plot(x.array, grad(relu)(x)[0].array, label="ReLU'(x)")
plt.plot(x.array, grad(grad(relu))(x)[0].array, label="ReLU''(x)")
plt.legend()
plt.show()
```

## Sigmoid
The `minijax.grad` function transformation can be nested arbitrarily.

```python
def sigmoid(x):
    return Array(1) / (Array(1) + exp(-x))
```

```python
plt.plot(x.array, sigmoid(x).array, label="Sigmoid(x)")
plt.plot(x.array, grad(sigmoid)(x)[0].array, label="Sigmoid'(x)")
plt.plot(x.array, grad(grad(sigmoid))(x)[0].array, label="Sigmoid''(x)")
plt.plot(x.array, grad(grad(grad(sigmoid)))(x)[0].array, label="Sigmoid'''(x)")
plt.plot(x.array, grad(grad(grad(grad(sigmoid))))(x)[0].array, label="Sigmoid''''(x)")
plt.plot(x.array, grad(grad(grad(grad(grad(sigmoid)))))(x)[0].array, label="Sigmoid'''''(x)")
plt.legend()
plt.show()
```
